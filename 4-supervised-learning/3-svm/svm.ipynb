```python
def validate_model(model, X, y):
    return cross_val_score(model, X, y=y, cv=10)
```

Remember that our decision boundary is defined by:

$$w^{T}x + b = 0$$

And in the case of a two feature dataset this is:

$$w_{1}X_{1} + w_{2}X_{2} + intercept = 0$$

And through some algebra our decision boundary is:

$$\frac{-w_{1}}{w_{2}}X_{1} - \frac{intercept}{w_{2}} = X_{2}$$

and more compactly

$$\alpha X_{1} - \frac{intercept}{w_{2}} = X_{2}$$

And for our margins it is very similar:

$$Minus \space plane = w_{1}X_{1} + w_{2}X_{2} + intercept = -1$$

$$Plus \space plane = w_{1}X_{1} + w_{2}X_{2} + intercept = +1$$

```python
def plot_decision(model, X, y, fig, ax, margins=False):
    # set limits
    xmin = np.min(X[:, 0]) - 0.2
    xmax = np.max(X[:, 0]) + 0.2
    ymin = np.min(X[:, 1]) - 0.2
    ymax = np.max(X[:, 1]) + 0.2

    # find decision boundary 
    model.fit(X, y)
    w = model.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(xmin,xmax, 1000)
    yy = a * xx - model.intercept_[0] / w[1]
    
    def plot_vector(ax1):
        # find support vectors and margins
        b = model.support_vectors_[0]
        minus_plane = a * xx + (b[1] - a * b[0])
        b = model.support_vectors_[-1]
        plus_plane = a * xx + (b[1] - a * b[0])
    
        # plot support vectors
        ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
                s=50, c='k')
        
        # plot margin
        ax.plot(xx, minus_plane, 'r--')
        ax.plot(xx, plus_plane, 'b--')
    
    
    if margins:
        plot_vector(ax)
        
    reg = ("; C = " + str(model.get_params()['C'])) if type(model) == SVC else ''
    # plot decision boundary
    ax.plot(xx, yy, label=str(type(model)).split(".")[-1][:-2] + reg)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)

    ax.legend()
    
    plt.xlim((xmin,xmax))
    plt.ylim((ymin, ymax))
    return fig, ax
```

```python
fig, ax = plt.subplots()
figt, axt = plot_decision(clf, X, y, fig, ax, margins=True)
```

```python
fig, ax = plt.subplots(figsize=(12,10))

rng = np.logspace(0, 5, 6) * 0.001

for c in rng[:3]:
    fig, ax = plot_decision(SVC(kernel="linear", C=c), X, y, fig, ax)
```

## Multiclass

```python
from sklearn.datasets import load_digits
digits = load_digits()

X = digits.data
y = digits.target

parameters = {'C':[0.0001,0.005, 0.1,10, 100, 1000], 'gamma':10.0 ** np.arange(-5, 4)}
est = SVC(kernel="rbf")
clf = GridSearchCV(est, parameters, cv=3)
clf.fit(X, y)
print clf.best_estimator_
print clf.best_score_
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.cross_validation import train_test_split

clf = SVC(kernel = 'rbf', C = 10 , gamma = 0.001 )

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=0)

clf.fit(X_train, y_train)

predicted = clf.predict(X_test)

confusion_matrix(y_test, predicted)
```
