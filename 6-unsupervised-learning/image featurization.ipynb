```python
def tranform_all_images(data_dir, shape, grayscale=True):
    # Make a new directory for the new transformed images
    if grayscale:
        new_dir_name = 'transformed_gray_%s_%s' % shape
    else: 
        new_dir_name = 'transformed_%s_%s' % shape
        
    if not os.path.exists(new_dir_name):
        os.mkdir(new_dir_name)

    # Loop through all the images transform and write to the directory
    for fname in os.listdir(data_dir):
        image = io.imread(os.path.join(data_dir, fname))
        if grayscale:
            image = color.rgb2gray(image) # gray scale
        transformed_image = transform.resize(image, shape)
        io.imsave(os.path.join(new_dir_name, fname), transformed_image)

tranform_all_images('data', (300, 300))
```
