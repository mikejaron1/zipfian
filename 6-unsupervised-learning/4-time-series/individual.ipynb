Time series can be in regular intervals or a fixed frequency; sec, min, hour, or it could be in irregular units or an unfixed frequency.
There are 4 types of time-series data that you will usually encounter: 

1. Timestamps, specific instants in time
2. Fixed periods, such as the month January 2007 or the full year 2010.
3. Intervals of time, indicated by a start and end timestamp. Periods can be thought of as special cases of intervals.
4. Experiment time: a measure of elapsed time with respect to 0 (the start of an experiment). We can convert any time series into an experiment time by setting one event as 0 and measuring all other time points as deltas from that point.

## Creating models in time series data: 

We typically want to decompose 4 components in a time series: 

1. Trend: are the data values increasing or decreasing? 
2. Seasonality: there is a distinct period or frequency that controls the values. 
3. Cyclic behavior: the mean of the data values changes over time resulting in no long term trend change. This is distinct from seasonality.

### Moving average window: 

The data as it stands has high variance, and it would be hard to make any predictions about trends in its current state. Let's go over a few methods we can use to smooth out short term fluctuations and highlight repetative features and cyclical activity.  There are 3 main types: 

1. Simple: a regular average is applied over the window. $T_t = \frac{1}{m}∑_{j=−k}^k \times \ y_{t+j} $
2. Cumulative: comparing the newest time point to the average of the whole time series. 
3. Weighted: Has multiplying factors to give different weights to data at different positions in the sample window.

Of the weighted moving averages, exponential weighted moving average (EWMA) is the most common. It is also known as [exponential smoothing](http://en.wikipedia.org/wiki/Exponential_smoothing) when applied to time series data. We assign a series of weights to each data point in relation to their 'age' in the window, with older points receiving less weight and more recent points (as defined by the window) recieving more weight. This has the effect of biasing the average towards newer data points and reducing the influence of older data points. 

![EWMA](http://upload.wikimedia.org/wikipedia/commons/b/ba/Exponential_moving_average_weights_N%3D15.png)

Here we introduce the `pd.rolling_main` and `pd.ewma` functions, which apply a rolling mean and exponential weighted moving average, respectively. Here we are averaging by week, denoted by the window size of 7 for the MA, and `span` of 7 in the EWMA. Notice that the rolling mean doesn't start until march. Why might this be?

```python
perday.plot()
data_avg_2 = pd.rolling_mean(perday, 30)
data_avg_2.index = data_avg_2.index - pd.offsets.Day(15)
data_avg_2.plot(label='RM-7')
data_ewma = pd.ewma(perday, span=30)
data_ewma.plot(label='EWMA')
plt.legend(loc=2)
```

## Hodrick-Prescott Filter
![HPFilter](http://upload.wikimedia.org/math/2/e/a/2eaa42f9977d2457c4650c4f7f80bfe2.png)
The first term of the equation is the sum of the squared deviations $d_t=y_t-\tau_t$ which penalizes the cyclical component. The second term is a multiple $\lambda$ of the sum of the squares of the trend component's second differences. This second term penalizes variations in the growth rate of the trend component. The larger the value of $\lambda$, the higher is the penalty. Hodrick and Prescott suggest 1600 as a value for $\lambda$ for quarterly data. Ravn and Uhlig (2002) state that $\lambda$ should vary by the fourth power of the frequency observation ratio; thus, $\lambda$ should equal 6.25 for annual data and $129,600$ for monthly data.

```python
from statsmodels.tsa import *
tsa = sm.tsa

'''
separates the cyclical component of the time series (here thought of as random noise) from the trend component.
the pickup in February is easier to observe at this scale. Should we call this a seasonal effect?
'''

tweet_cycle, tweet_trend = tsa.filters.hpfilter(t400, 10000)

tweet_decomp = pd.DataFrame(t400)
tweet_decomp['cycle'] = tweet_cycle
tweet_decomp['trend'] = tweet_trend

tweet_decomp.plot()
```

```python
# Imports
from sklearn import svm
from sklearn.ensemble import RandomForestRegressor
from sklearn import grid_search
from sklearn import metrics
from sklearn import preprocessing
from sklearn import cross_validation

# Preprocessing
def extract_date_feature(df):
    df['hour'] = df.index.map(lambda idx: idx.hour)
    df['weekday'] = df.index.map(lambda idx: idx.weekday())
    
    encoder = preprocessing.OneHotEncoder()
    encoder.fit_transform(df[['hour','weekday']])
    
    encoded = pd.concat([pd.get_dummies(df[col], prefix=col) for col in df.columns if not col in ['demand']], axis=1)
    return (pd.concat([df, encoded], axis=1), encoder)

def vectorize_date_feature(date, encoder):
    
    datestamp = pd.to_datetime(date)
    return encoder.transform([[datestamp.hour,datestamp.weekday()]]).toarray()[0]

# Encode the continuous data
dataset, encoder = extract_date_feature(df)

# We only want to train on encoded data
X_cols = [col for col in dataset.columns if not col in [u'demand', u'hour', u'weekday']]
y_cols = 'demand'
```

```python
# Split the dataset into train and test sets
split = int(dataset.shape[0] * 0.5)

# Get the test and training sets into a format that scikit-learn likes
X, y = dataset[X_cols], dataset[y_cols]
(X_train, X_test, y_train, y_test,) = X[0:split], X[split:], y[0:split], y[split:]
# (X_train, X_test, y_train, y_test,) = cross_validation.train_test_split(X,y)

params = {
    'C': [1,10,100],
    'kernel': ['rbf','sigmoid'],
    'gamma': [0.0, 0.1],
}

params_rf = {
    'n_estimators' : [1, 10, 100, 100],
    'max_features' : [1, 5, 10, 20, 30],
    'max_depth' : [1, 3, 5, 7, 10]
}

optimizer = grid_search.GridSearchCV(svm.SVR(), param_grid=params, verbose=True)
optimizer.fit(X_train, y_train)
best_params = optimizer.best_params_;
model = optimizer.best_estimator_
```

```python
# Check the performance of out of sample predictions ( on the test dataset )
prediction = model.predict(X_test)
print("Out of sample MSE:{0}".format(metrics.mean_squared_error(y_test,prediction)))
plt.plot(X_test.index, prediction, label='prediction')
plt.plot(y_test.index, y_test, lw=0.2, color='k',label='real')
plt.legend()

# Compare against the performance of the model on the training dataset.
in_sample_prediction = model.predict(X_train)
print("In sample MSE: {0}".format(metrics.mean_squared_error(y_train,in_sample_prediction)))
plt.plot(X_train.index, in_sample_prediction, label='in_sample_prediction')
plt.plot(y_train.index, y_train, lw=0.2, color='k',label='real',)
plt.legend()

# Sanity check, check how much better we're doing than a mean-estimator
from sklearn import dummy
r = dummy.DummyRegressor()
r.fit(X_train,y_train)
dummy_prediction = r.predict(X_test)
print("Dummy MSE: {0}".format(metrics.mean_squared_error(y_test,dummy_prediction)))
plt.plot(X_test.index, dummy_prediction, label='dummy_prediction')
plt.legend()
```

```python
# Autocorrelation
import statsmodels.api as sm
from statsmodels.graphics import tsaplots

nlags=14*24
correlation_threshold = .5
acf = sm.tsa.acf(df['demand'], nlags=nlags)
acf = pd.Series(data=acf, index=np.arange(nlags+1), name='acf')
lags = acf[acf >= correlation_threshold].index
print("Lags with highest autocorrelation {l}".format(l=lags))

ax = acf.plot()
ymin, ymax = ax.get_ylim()
high_acf = acf.index.map(lambda x: x in lags)
ax.fill_between(acf.index,correlation_threshold, acf,where=high_acf,alpha=1,color='r',label='high acf')
plt.legend()
```

```python
import statsmodels.api as sm
from sklearn import metrics

# Train
maxlag=14*24
split = int(df.shape[0] * .8)
train, test = df['demand'][0:split], df['demand'][split:]

model = sm.tsa.AR(endog=train).fit(maxlag=maxlag)

# In sample Prediction
start = train.index[model.k_ar].isoformat()
end = train.index[-1].isoformat()
rng = pd.date_range(start=start,end=end,freq='H')

in_sample_prediction = model.predict(start=start,end=end,dynamic=True)

print("In sample MSE: {0}".format(metrics.mean_squared_error(df['demand'][rng], in_sample_prediction)))
in_sample_prediction.plot(label='in_sample_prediction')

# Out of sample Predict
start = train.index[-1].isoformat()
end = test.index[-1].isoformat()
rng = pd.date_range(start=start,end=end,freq='H')

prediction = model.predict(start=start,end=end,dynamic=True)

print("Out of sample MSE: {0}".format(metrics.mean_squared_error(df['demand'][rng], prediction)))
prediction.plot(label='prediction')
df['demand'].plot(lw=.2)
plt.legend()
```
