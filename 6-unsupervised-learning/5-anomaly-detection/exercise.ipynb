#### We can model page views with a Poisson since a Poisson models count data (for a given time period) with a paramter (lambda) equal to the average.

```python
# scale data
page_views = df['page-views'] / 10000.
```

```python
df['page-views'].mean()
```

```python
page_views.plot()
```

```python
# Method of moments (also happens to be MLE for Poisson)
# http://en.wikipedia.org/wiki/Poisson_distribution#Maximum_likelihood
lam = page_views.mean()
```

```python
page_views.hist()
```

```python
import scipy.stats as sp

lam
```

```python
x = np.arange(10)
hist, bins = np.histogram(page_views, bins=20)
```

```python
hist, bins
```

```python
plt.bar(bins[:-1], hist.astype(np.float32) / hist.sum(), width=(bins[1]-bins[0]))
plt.plot(sp.poisson.pmf(x, lam, loc=-1), color="red")
```

```python
import math
lambdas = np.linspace(0,2)
poisson_like = lambda x, lam: np.exp(-lam) * (lam**x) / (np.arange(x)+1).prod()


log_likelihood = []

for l in lambdas:
    sums = 0
    for yi in page_views:
        likeli = poisson_like(yi, l)
        sums += 0 if likeli == 0 else math.log(likeli)
    log_likelihood.append(sums)

plt.plot(lambdas, log_likelihood)
plt.title("Log Likelihood of a Poisson")
plt.xlabel('$\lambda$')
plt.ylabel('Log(L($\lambda$|x))')
```

```python
np.argmax(log_likelihood[1:])
```

## Phony Domains

```python
import sklearn.feature_extraction
import pandas as pd
import sklearn.cross_validation

alexa_dataframe = pd.read_csv('data/alexa.csv', names=['rank','uri'], header=None, encoding='utf-8')
alexa_dataframe.head()
```

```python
import tldextract

def domain_extract(uri):
    ext = tldextract.extract(uri)
    if (not ext.suffix):
        return np.nan
    else:
        return ext.domain

alexa_dataframe['domain'] = [ domain_extract(uri) for uri in alexa_dataframe['uri']]
del alexa_dataframe['rank']
del alexa_dataframe['uri']
alexa_dataframe.head()
```

```python
alexa_dataframe.tail()
```

```python
alexa_dataframe = alexa_dataframe.dropna()
alexa_dataframe = alexa_dataframe.drop_duplicates()

alexa_dataframe['class'] = 'real'

print 'Number of Alexa domains: %d' % alexa_dataframe.shape[0]
```

```python
alexa_dataframe.head()
```

```python
fake_dataframe = pd.read_csv('data/spoofs.csv', names=['raw_domain'], header=None, encoding='utf-8')

# We noticed that the blacklist values just differ by captilization or .com/.org/.info
fake_dataframe['domain'] = fake_dataframe.applymap(lambda x: x.split('.')[0].strip().lower())
del fake_dataframe['raw_domain']

# It's possible we have NaNs from blanklines or whatever
fake_dataframe = fake_dataframe.dropna()
fake_dataframe = fake_dataframe.drop_duplicates()
fake_total = fake_dataframe.shape[0]
print 'Total DGA domains %d' % fake_total

# Set the class
fake_dataframe['class'] = 'fake'

print 'Number of DGA domains: %d' % fake_dataframe.shape[0]
```

```python
fake_dataframe.head()
```

```python
all_domains = pd.concat([alexa_dataframe, fake_dataframe], ignore_index=True)
```

```python
all_domains.head()
```

```python
all_domains['length'] = [len(x) for x in all_domains['domain']]
```

```python
import math
from collections import Counter
 
def entropy(s):
    p, lns = Counter(s), float(len(s))
    return -sum( count/lns * math.log(count/lns, 2) for count in p.values())

all_domains['entropy'] = [entropy(x) for x in all_domains['domain']]
```

## Supervised Learning

```python
X1 = all_domains.as_matrix(['length', 'entropy'])
y1 = np.array(all_domains['class'].tolist())
```

```python
import sklearn.ensemble
clf = sklearn.ensemble.RandomForestClassifier(n_estimators=20)
```

```python
scores = sklearn.cross_validation.cross_val_score(clf, X1, y1, cv=5, n_jobs=4)
print scores
```

```python
# Wow 97% accurate! At this point we could claim success but we would be foolish...
# Recall that we have ~100k 'real' domains and only 3.5k fake domains
# So a classifier that marked everything as real would be about
# 96% accurate....

# So we dive in a bit and look at the predictive performance more deeply.
X_train1, X_test1, y_train1, y_test1 = sklearn.cross_validation.train_test_split(X1, y1, test_size=0.2)
clf.fit(X_train1, y_train1)
y_pred1 = clf.predict(X_test1)
```

```python
labels1 = ['real', 'fake']
cm1 = confusion_matrix(y_test1, y_pred1, labels1)

def plot_cm(cm, labels):
    
    # Compute percentanges
    percent = (cm*100.0)/np.array(np.matrix(cm.sum(axis=1)).T)
    
    print 'Confusion Matrix Stats'
    for i, label_i in enumerate(labels):
        for j, label_j in enumerate(labels):
            print "%s/%s: %.2f%% (%d/%d)" % (label_i, label_j, (percent[i][j]), cm[i][j], cm[i].sum())

    # Show confusion matrix
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.grid(b=False)
    cax = ax.matshow(percent, cmap='coolwarm')
    pylab.title('Confusion matrix of the classifier')
    fig.colorbar(cax)
    ax.set_xticklabels([''] + labels)
    ax.set_yticklabels([''] + labels)
    pylab.xlabel('Predicted')
    pylab.ylabel('True')
    pylab.show()

plot_cm(cm1, labels1)
```

## Unsupervised

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
import random as rnd

data = all_domains[['length', 'entropy', 'class']]
rows = rnd.sample(data.index, 5000)
data = data.ix[rows]
X = data[['length', 'entropy']]
X = StandardScaler().fit_transform(X.as_matrix())
labels_true = data['class'].tolist()
```

```python
db = DBSCAN(eps=0.7, min_samples=10).fit(X)

##############################################################################
# Plot result
import pylab as pl

def plot_dbscan(X, db, true_labels):
    core_samples = db.core_sample_indices_
    labels = db.labels_
    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    
    print('Estimated number of clusters: %d' % n_clusters_)
    print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
    print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
    print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
    print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(labels_true, labels))
    print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(labels_true, labels))
    print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels))

    # Black removed and is used for noise instead.
    unique_labels = set(labels)
    colors = pl.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = 'k'
            markersize = 6
        class_members = [index[0] for index in np.argwhere(labels == k)]
        cluster_core_samples = [index for index in core_samples
                                if labels[index] == k]
        for index in class_members:
            x = X[index]
            stroke='k'
            if index in core_samples and k != -1:
                markersize = 14
            else:
                markersize = 6
                
            stroke = 'r' if true_labels[index] == 'fake' else 'k'
            
            pl.plot(x[0], x[1], 'o', markerfacecolor=col,
                    markeredgecolor=stroke, markersize=markersize)

    pl.title('Estimated number of clusters: %d' % n_clusters_)


# Fake domains are outlined in red    
plot_dbscan(X, db, labels_true)
```
